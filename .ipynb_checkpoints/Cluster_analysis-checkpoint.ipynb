```python
#import modules
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
```

```python
#import data
df = pd.read_csv("MOCAP_ALL_INDIVIDUAL_SWINGS.csv")
print(f"Data Size: {df.shape}")
backup_data = df[:]
```

### FEATURE REDUCTION

```python
#get rid of all player names with integers in them
df['Player'] = df['Player'].str.replace('\d+', '')
```

```python
#dealing with NaN values

#1 find how many NaN values in each row
NaN_series = df.isnull().sum(axis=1).sort_values(ascending=False)

#find a good threshold of rows to just delete completely
#final: delete all rows with more than 10 NaN values (coverage too poor)
z = 0
thresh = 10
bad_rows = NaN_series[(NaN_series > z) == True].shape
worst_rows = NaN_series[(NaN_series > thresh) == True].shape
print(f"Number of Rows w NaN Values: {bad_rows[0]}")
print(f"Number of Rows w more than {thresh} NaN Values: {worst_rows[0]}")
```

```python
#delete rows that have NaN values above this threshold
data_drop = NaN_series[(NaN_series > thresh) == True].index
df = df.drop(df.index[data_drop])
print(df.shape)
```

```python
#fill NaN values with approximated values
column_names = list(df)
for i in range(1,len(df.columns)):
    if (df.iloc[:,i].isna().any() == True):
        x = df.groupby("Player")[column_names[i]].transform('median')
        #fill empty rows with the corresponding median value in this pandas series
        df[column_names[i]].fillna(x, inplace=True)
```

```python
#fill NaN values with median values of that player
column_names = list(df)
if (df.iloc[:,i].isna().any() == True):
    x = df.groupby("Player")[column_names[i]].transform('median')
    #fill empty rows with the corresponding median value in this pandas series
    df[column_names[i]].fillna(x, inplace=True)
```

```python
#show new completed data
print(df.shape)
print(f"number of players: {len(np.unique(df['Player']))}")
print(f"number of sessions total: {len(df['Player'])}")
df.head()
```

### Random Undersample Data due to Imbalance

```python
#Randomly oversample players that have less than 30 swings
#This is because we want a baseline of 30 swings for undersampling
for class_index, group in df.groupby('Player'):
    if len(group)<30:
        df = df.append(group.sample(30-len(group),replace=True))
```

```python
X = df.iloc[:,1:]
y = df['Player']
```

```python
from imblearn.under_sampling import RandomUnderSampler

#Undersampling
rus = RandomUnderSampler(random_state=2)
rus.fit(X, y)
X_under, y_under = rus.fit_sample(X, y)
```

```python
#Let's use undersampling data
df = pd.DataFrame(X_under)
df.columns = list(X)
df['Player'] = y_under
print(df.shape)
df.head(100)
```

```python
#now you can see that every player equally has 30 swings to contribute
df['Player'].value_counts()
```

### Hierarchical Clustering

```python
#create new data set for clustering/scaling
df_clust = df[:]

del df_clust['headID']
del df_clust['Player']

#store column names
columns = list(df_clust)

#convert to float
df_clust = df_clust.astype(float)

#standardize the data
scaler = StandardScaler()

#fit the scaler to our training set
df_scaled = pd.DataFrame(scaler.fit_transform(df_clust))

#input columns
df_scaled.columns = columns
df_scaled.head()
```

```python
# import hierarchical clustering libraries
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
# calculate full dendrogram
plt.figure(figsize=(30, 20))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('sample index')
plt.ylabel('distance')
Z = linkage(df_scaled, method='ward')
dendrogram(
    Z,
    truncate_mode='lastp',  # show only the last p merged clusters
    p=6,  # show only the last p merged clusters
    leaf_rotation=90.,  # rotates the x axis labels
    leaf_font_size=20.,  # font size for the x axis labels
    show_contracted=True,
)
plt.show()
```

```python
# create clusters based on dendrogram above
hc = AgglomerativeClustering(n_clusters=4, affinity = 'euclidean', linkage = 'ward')
# save clusters for chart
y_hc = hc.fit_predict(df_scaled)
```

```python
#test visualize clusters
plt.figure(figsize=(10, 7))  
plt.scatter(df_scaled.iloc[:,0], df_scaled.iloc[:,11], c=hc.labels_, cmap='rainbow')
```

```python
#Append clusters to original data for analysis
df.insert(2, column='SwingType', value=y_hc)
```
